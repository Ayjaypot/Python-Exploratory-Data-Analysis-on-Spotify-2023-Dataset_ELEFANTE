# Basic Descriptive Statistics
### - What are the mean, median, and standard deviation of the streams column?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The mean, median, and mode can be seen below before the graph.
### - What is the distribution of released_year and artist_count? Are there any noticeable trends or outliers?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The distribution of the released year and artist count can be seen in the graph. As for the trends in released year there are noticeably less tracks from 2000 and below, &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; and there is an almost steady increase for the songs 2000 up until 2023. As for the artist count the number of tracks goes down as the number of artists increases.eases.

```python
#mean, median, and standard deviation
#converts other data types to integers
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
#calculates the mean
mean_value = df['streams'].mean()
print('mean:', mean_value)
#calculates the median
median_value = df['streams'].median()
print('Median:', median_value)
#calculates the mode
mode_value = df['streams'].mode()
print('Mode:', "\n", mode_value) 

#graph for distribution of released_year 
plt.figure(figsize=(10, 6))
sns.histplot(df['released_year'], bins=100,)  
plt.title('Distribution of Released Year')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.show()

#graph for distribution of artist_count
plt.figure(figsize=(10, 6))
sns.histplot(df['artist_count'], bins=100,)  
plt.title('Distribution of Artist Count')
plt.xlabel('Artist Count')
plt.ylabel('Frequency')
plt.show()
```

# Temporal Trends
### - Analyze the trends in the number of tracks released over time. Plot the number of tracks released per year.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Same with the earlier graph the tracks released per year almost undergoes a steady increase with the years between 2000 and 2020 having the most number of tracks.

### - Does the number of tracks released per month follow any noticeable patterns? Which month sees the most releases?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; With the number of tracks released per month, the month with the most releases is January and there is no noticeable trend except the some pair of months have almost the same number of releases.

```python
# sorts the dataframe by the counts of the released year
tracks_per_year = df['released_year'].value_counts().sort_index()

# Plot the number of tracks released per year 
plt.figure(figsize=(10, 6))
plt.bar(tracks_per_year.index, tracks_per_year.values, color='yellow')
plt.title("Number of Tracks Released Per Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Tracks Released", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=1)
plt.show()

# Plot the number of tracks released per month
tracks_per_month = df['released_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(tracks_per_month.index, tracks_per_month.values, color='lightblue')
plt.title("Number of Tracks Released Per Month")
plt.xlabel("Month")
plt.ylabel("Number of Tracks Released")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(axis='y', linestyle='--', alpha=1)
plt.show()
```

# Genre and Music Characteristics
### - Examine the correlation between streams and musical attributes like bpm, danceability_%, and energy_%. Which attributes seem to influence streams the most?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; As shown with the computed data below almost all have a negative correlation with bpm having almost no correlation which means that none of this attributes affect the streams much.

### - Is there a correlation between danceability_% and energy_%? How about valence_% and acousticness_%?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; There is a positive correlation between danceablity and energy which means they go up in relation to each other, while the opposite is true for valence and acousticness

```python
#setting a new dataframe which only includes specific columns
columns = ['streams', 'bpm', 'danceability_%', 'energy_%']
spotify_df = df[columns]

# Calculate correlation 
correlation = spotify_df.corr()

# Correlation between the needed columns
correlation_with_streams = correlation['streams'].drop('streams').sort_values(ascending=False)
print("Correlation with streams:","\n", correlation_with_streams)
print("\n")

#setting a new dataframe which only includes specific columns
new_columns = ['danceability_%', 'energy_%', 'valence_%', 'acousticness_%']
new_spotify_df = df[new_columns]

# Calculate the correlation for these specific columns
new_correlation = new_spotify_df.corr()
danceability_energy_corr = new_correlation.loc['danceability_%', 'energy_%']
valence_acousticness_corr = new_correlation.loc['valence_%', 'acousticness_%']

print("Correlation between danceability_% and energy_%:", danceability_energy_corr)
print("Correlation between valence_% and acousticness_%:", valence_acousticness_corr)
```

# Platform Popularity
### - How do the numbers of tracks in spotify_playlists, in_deezer_playlists, and apple_playlists compare? Which platform seems to favor the most popular tracks?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; As can be seen in the total and average in playlists by platform, in spotify playlists favor the most popular tracks.

```python
#setting the dataframe for the specific columns
playlist_columns = ['in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']
df[playlist_columns] = df[playlist_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Calculate total and average playlist counts per platform asnd prints it
total_playlists = df[playlist_columns].sum()
average_playlists = df[playlist_columns].mean()

print("Total Tracks in Playlists by Platform:","\n", total_playlists)
print("\n")
print("Average Tracks in Playlists by Platform:","\n", average_playlists)



```

# Advanced Analysis
### - Based on the streams data, can you identify any patterns among tracks with the same key or mode (Major vs. Minor)?
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The streams for tracks with the same key are almost the same with each other the same with the major or minor mode.
### - Do certain genres or artists consistently appear in more playlists or charts? Perform an analysis to compare the most frequently appearing artists in playlists or charts.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; An analysis of the most frequent artists shows the top 5 below.

```python
#Advanced Analysis
# Group by key and calculate average streams for each key
average_streams_by_key = df.groupby('key')['streams'].mean()

# Group by mode and calculate average streams for Major and Minor modes and prints it
average_streams_by_mode = df.groupby('mode')['streams'].mean()

print("Average Streams by Key:","\n",average_streams_by_key)
print("\n")
print("Average Streams by Mode (Major vs. Minor):","\n",average_streams_by_mode)
print("\n")

# Convert columns needed to numbers
playlist_chart_columns = [
    'in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists',
    'in_apple_charts', 'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts'
]

df[playlist_chart_columns] = df[playlist_chart_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Calculate total appearances in playlists and charts for each artist
df['total_playlist_chart_appearances'] = (
    df['in_spotify_playlists'] +
    df['in_spotify_charts'] +
    df['in_apple_playlists'] +
    df['in_apple_charts'] +
    df['in_deezer_playlists'] +
    df['in_deezer_charts'] +
    df['in_shazam_charts']
)

# groups by artist and finds the total playlist and chart appearances
artist_appearances = df.groupby('artist(s)_name')['total_playlist_chart_appearances'].sum()

# prints the top 5 artists
top_artists = artist_appearances.sort_values(ascending=False).head(5)
print("Top 5 Artists by Playlist and Chart Appearances:", "\n", top_artists)
```
